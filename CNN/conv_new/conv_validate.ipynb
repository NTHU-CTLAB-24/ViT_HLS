```python
import torch

x = torch.zeros(2, 3, 16, 16)

for n in range(2):
    for c in range(3):
        for h in range(16):
            for w in range(16):
                x[n, c, h, w] = h + c

kernel = torch.zeros((2, 3, 3, 3))
for k in range(2):
    for l in range(3):
        for i in range(3):
            for j in range(3):
                kernel[k, l, i, j] = j + k

conv = torch.nn.Conv2d(in_channels=3, out_channels=2,
                       kernel_size=3, bias=False, stride=1, padding=0)
conv.weight.data = kernel

res = conv(x)

print(res)
```
